```python
#Create Year column, it does not need lambda function
cleaned_properties['Year'] = cleaned_properties['Month'].dt.year
```

```python
#Check if year column is created
cleaned_properties.head()
```

```python
avg_yr = cleaned_properties.groupby(by = ['London_Borough', 'Year'])[['Average_price']].mean()
```

```python
idx = pd.IndexSlice
```

```python
avg_yr.columns
```

```python
london_2018 = avg_yr.loc[idx[:, 2018],:]
london_1998 = avg_yr.loc[idx[:, 1998],:]
```

```python
two_decade = london_2018.reset_index()[['London_Borough']]
```

```python
two_decade['Price_change'] = london_2018.reset_index()['Average_price'] - london_1998.reset_index()['Average_price']
```

```python
two_decade = two_decade.sort_values('Price_change', ascending = False)
```

```python
# Which London Borough average price has increased most in 20 years
# 2018 - 1998
plt.barh('London_Borough', 'Price_change', data = two_decade.head(10))
plt.title('Top 10 price change in 20 year')
plt.xlabel('Price Change')
plt.ylabel('London Borough');
print(two_decade[two_decade['Price_change'] == two_decade['Price_change'].max()])
```
